```python
# Import the pipeline function
from transformers import pipeline
import warnings

# Suppress specific warnings from the models loading, optional but cleaner output
warnings.filterwarnings("ignore", category=UserWarning, module='transformers.modeling_utils')

print("Loading models... This might take a minute or two.")

# Load a pipeline for Toxicity Detection
# This model might output multiple labels (toxic, severe_toxic, etc.) or just 'toxic'/'non-toxic'.
# We'll check its output structure. Using 'return_all_scores=True' helps see the model's confidence.
try:
    # Trying a multi-label toxicity model first
    tox_classifier = pipeline(
        "text-classification",
        model="unitary/toxic-bert", # A common model for toxicity
        return_all_scores=True
    )
    print("Using unitary/toxic-bert model for toxicity.")
    MULTI_LABEL_TOXICITY = True
except Exception as e:
    print(f"Failed to load unitary/toxic-bert, trying another toxicity model: {e}")
    # Fallback to a simpler binary toxic/non-toxic model if the first one fails
    tox_classifier = pipeline(
        "text-classification",
        model="martin-ha/toxic-comment-model", # Another option
        return_all_scores=True
    )
    print("Using martin-ha/toxic-comment-model for toxicity.")
    MULTI_LABEL_TOXICITY = False


# Load a pipeline for Sentiment Analysis
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-roberta-base-sentiment-latest", # A robust sentiment model
    top_k=None # Return scores for all labels (positive, negative, neutral)
)

print("Models loaded successfully!")
```

```python
# Define a threshold for considering toxicity significant
# You can adjust this - higher means less sensitive, lower means more sensitive.
TOXICITY_THRESHOLD = 0.5

def analyze_text_bias(text):
    """
    Analyzes input text for potential toxicity and sentiment bias indicators.
    """
    print(f"\n--- Analyzing Text: \"{text}\" ---")
    results = {
        "potential_toxicity": [],
        "sentiment_analysis": None,
        "potential_framing_bias": None,
        "warnings": []
    }

    # --- 1. Toxicity Analysis ---
    try:
        toxicity_scores = tox_classifier(text)[0] # Get scores for the input text

        print("\nDetailed Toxicity Scores:")
        for score_info in toxicity_scores:
             print(f"  - {score_info['label']}: {score_info['score']:.4f}")

        # Check which toxicity labels exceed the threshold
        if MULTI_LABEL_TOXICITY:
            # For models like toxic-bert with multiple labels
             for score_info in toxicity_scores:
                 if score_info['score'] > TOXICITY_THRESHOLD:
                     results["potential_toxicity"].append(f"{score_info['label']} (Score: {score_info['score']:.2f})")
        else:
             # For models that might have simpler output (e.g., just 'toxic' or specific labels)
             # Find the label with the highest score, often 'toxic' or 'LABEL_1' etc.
             # We assume the label indicating toxicity is the primary one we care about.
             # This part might need adjustment based on the specific fallback model's output labels.
             # Let's check for common 'toxic' labels or the label with highest score if generic.
             toxic_label_found = False
             for score_info in toxicity_scores:
                 label_lower = score_info['label'].lower()
                 # Check if the label itself suggests toxicity and score is high
                 if ('toxic' in label_lower or 'insult' in label_lower or 'obscene' in label_lower or 'threat' in label_lower or 'hate' in label_lower) and score_info['score'] > TOXICITY_THRESHOLD:
                     results["potential_toxicity"].append(f"Potential {score_info['label']} (Score: {score_info['score']:.2f})")
                     toxic_label_found = True
             # If no specific toxic label found, check the highest score if it seems relevant
             # (This is a heuristic and might misinterpret generic 'LABEL_1' etc.)
             if not toxic_label_found:
                 highest_score_item = max(toxicity_scores, key=lambda x: x['score'])
                 # Be cautious flagging generic labels unless score is very high
                 if highest_score_item['score'] > 0.75: # Higher threshold for generic labels
                      results["warnings"].append(f"High score for generic label '{highest_score_item['label']}' ({highest_score_item['score']:.2f}). This *might* indicate toxicity, needs review.")


        if not results["potential_toxicity"] and not any("High score for generic label" in w for w in results["warnings"]):
             print("\nResult: Low indication of toxicity found.")
        else:
             print(f"\nResult: Potential Toxicity Indicators Found: {results['potential_toxicity']}")
             if any("High score for generic label" in w for w in results["warnings"]):
                 print(f"Warning: {results['warnings']}")


    except Exception as e:
        print(f"\nError during toxicity analysis: {e}")
        results["warnings"].append("Toxicity analysis failed.")

    # --- 2. Sentiment Analysis (Proxy for Framing) ---
    try:
        sentiment_scores = sentiment_analyzer(text)[0] # Get list of dicts for each label

        print("\nDetailed Sentiment Scores:")
        sentiment_label = ""
        highest_score = 0
        for score_info in sentiment_scores:
             print(f"  - {score_info['label'].capitalize()}: {score_info['score']:.4f}")
             if score_info['score'] > highest_score:
                 highest_score = score_info['score']
                 sentiment_label = score_info['label'].capitalize() # e.g., Positive, Negative, Neutral

        results["sentiment_analysis"] = f"{sentiment_label} (Score: {highest_score:.2f})"
        print(f"\nOverall Sentiment: {results['sentiment_analysis']}")

        # --- 3. Interpret Sentiment for Framing Bias ---
        # Framing bias is about *how* something is presented to influence perception.
        # Strong sentiment (very positive or very negative) can be an indicator.
        if sentiment_label == "Negative" and highest_score > 0.7:
             results["potential_framing_bias"] = "Strong negative sentiment detected. This *could* indicate negative framing bias, depending on the context. Check if the language aims to create an unfavorable view."
        elif sentiment_label == "Positive" and highest_score > 0.7:
            results["potential_framing_bias"] = "Strong positive sentiment detected. This *could* indicate positive framing bias, depending on the context. Check if the language aims to create an overly favorable view."
        else:
             results["potential_framing_bias"] = "Neutral or moderate sentiment detected. Less likely to indicate strong framing bias based *solely* on sentiment."

        print(f"Framing Bias Indication (based on sentiment): {results['potential_framing_bias']}")

    except Exception as e:
        print(f"\nError during sentiment analysis: {e}")
        results["warnings"].append("Sentiment analysis failed.")

    # --- 4. Final Summary ---
    print("\n--- Analysis Summary ---")
    if results["potential_toxicity"]:
        print(f"🔴 Potential Toxicity/Harmful Content Detected:")
        for item in results["potential_toxicity"]:
            print(f"   - {item}")
        print("   (Note: This can overlap with harmful stereotypes or attacks.)")
    else:
        print("✅ No significant toxicity indicators detected by the model.")

    if results["sentiment_analysis"]:
        print(f"\n🔵 Sentiment: {results['sentiment_analysis']}")
        print(f"   {results['potential_framing_bias']}")
    else:
        print("\n⚪ Sentiment analysis could not be completed.")

    if results["warnings"]:
        print("\n⚠️ Warnings during analysis:")
        for warning in results["warnings"]:
            print(f"   - {warning}")

    print("\n--- End of Analysis ---")
    return results
```

```python
# Example Usage

# Example 1: Potentially toxic stereotype
text1 = "All politicians are corrupt liars and can't be trusted."
analyze_text_bias(text1)

# Example 2: Potentially negative framing
text2 = "The proposed welfare cuts will devastate vulnerable families."
analyze_text_bias(text2)

# Example 3: Potentially positive framing (using sentiment)
text3 = "This groundbreaking tax relief plan will boost the economy for everyone!"
analyze_text_bias(text3)

# Example 4: More neutral statement
text4 = "The council meeting discussed the budget for the upcoming fiscal year."
analyze_text_bias(text4)

# Example 5: Subtle stereotype (might be missed or flagged mildly)
text5 = "She's very emotional, like most women are."
analyze_text_bias(text5)


# Get input from the user
print("\n\n=== Try your own text! ===")
user_text = input("Enter the text you want to analyze for bias indicators: ")
analyze_text_bias(user_text)
```
